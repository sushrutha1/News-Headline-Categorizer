```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datasets import load_dataset

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, confusion_matrix
```

```python
################################################################## Load AG News dataset from HuggingFace######################################
dataset = load_dataset("ag_news")

# Convert to pandas DataFrame
train_df = pd.DataFrame(dataset['train'])
test_df = pd.DataFrame(dataset['test'])

print(train_df.head())
print(train_df['label'].value_counts())
```

```python
# Features & Labels
X_train, y_train = train_df['text'], train_df['label']
X_test, y_test = test_df['text'], test_df['label']

# TF-IDF Vectorization
vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

```python
# Train Naive Bayes classifier
model = MultinomialNB()
model.fit(X_train_tfidf, y_train)

# Predictions
y_pred = model.predict(X_test_tfidf)
```

```python
# Classification Report
print(classification_report(y_test, y_pred, target_names=["World", "Sports", "Business", "Sci/Tech"]))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=["World", "Sports", "Business", "Sci/Tech"],
            yticklabels=["World", "Sports", "Business", "Sci/Tech"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
from sklearn.metrics import accuracy_score

# Calculate Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.4f}")
```

```python
import numpy as np

# Get feature names
feature_names = np.array(vectorizer.get_feature_names_out())

# Get log probabilities of features
class_labels = ["World","Sports","Business","Sci/Tech"]

for i, label in enumerate(class_labels):
    top10 = model.feature_log_prob_[i].argsort()[-10:]
    print(f"\nTop words for {label}: {feature_names[top10]}")
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Reduce TF-IDF vectors to 3D
pca = PCA(n_components=3, random_state=42)
X_train_3d = pca.fit_transform(X_train_tfidf.toarray()[:2000])  # use subset (2000) for speed
y_subset = y_train[:2000]

# Plot 3D Scatter
fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(
    X_train_3d[:,0], X_train_3d[:,1], X_train_3d[:,2],
    c=y_subset, cmap="tab10", alpha=0.6
)

# Axis labels
ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_zlabel("PC3")
ax.set_title("3D PCA Visualization of News Headlines")

# Legend
legend_labels = ["World","Sports","Business","Sci/Tech"]
plt.legend(handles=scatter.legend_elements()[0], labels=legend_labels)
plt.show()
```

```python
from sklearn.manifold import TSNE

# Use a subset for speed (t-SNE is heavy)
subset_size = 2000
X_subset = X_train_tfidf[:subset_size].toarray()
y_subset = y_train[:subset_size]

# Apply t-SNE (2D)
tsne = TSNE(n_components=2, random_state=42, perplexity=30, n_iter=1000)
X_2d = tsne.fit_transform(X_subset)

# Plot
plt.figure(figsize=(10,7))
sns.scatterplot(x=X_2d[:,0], y=X_2d[:,1], hue=y_subset, palette="tab10", alpha=0.7)
plt.title("t-SNE Visualization of News Headline Categories")
plt.legend(labels=["World","Sports","Business","Sci/Tech"])
plt.show()
```

```python
sample_headlines = [
    "Tata share values drops by 20 percent",
    "mu sigma launches carrer pad 2.0",
    "virat hit a six in this first test match",
    "meta opens  new office in mysuru"
]

sample_tfidf = vectorizer.transform(sample_headlines)
preds = model.predict(sample_tfidf)

for text, label in zip(sample_headlines, preds):
    print(f"{text} --> {['World','Sports','Business','Sci/Tech'][label]}")
```

```python
sample_headlines = [
    "Apple releases new iPhone with advanced camera",           # Sci/Tech
    "Microsoft announces breakthrough in AI research",          # Sci/Tech
    "Stock markets hit record high after tech rally",           # Business
    "Inflation rises as oil prices climb",                      # Business
    "Brazil wins the World Cup after thrilling final",          # Sports
    "Serena Williams retires after historic tennis career",     # Sports
    "United Nations discusses global climate change policies",  # World
    "chiefminister  addresses state after peace agreement"          # World
]

expected_labels = [
    "Sci/Tech","Sci/Tech","Business","Business",
    "Sports","Sports","World","World"
]

sample_tfidf = vectorizer.transform(sample_headlines)
preds = model.predict(sample_tfidf)

for text, pred, exp in zip(sample_headlines, preds, expected_labels):
    predicted_label = ["World","Sports","Business","Sci/Tech"][pred]
    print(f"Headline: {text}")
    print(f"Expected: {exp} | Predicted: {predicted_label}")
    print("-"*70)
```
